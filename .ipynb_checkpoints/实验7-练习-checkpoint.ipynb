# 5ï¸âƒ£æœºå™¨å­¦ä¹ è§†è§’ï¼šå­¦ä¹  $p(y|x)$ çš„ç®€å•æ¼”ç¤º

å‰ 20 é¡µè®²åˆ°æœºå™¨å­¦ä¹ çš„æœ¬è´¨æ˜¯å­¦ä¹ æ¡ä»¶åˆ†å¸ƒï¼š

$$

æˆ‘ä»¬æ„é€ ä¸€ä¸ªéå¸¸ç®€å•çš„ä¾‹å­ï¼š

- $x = 0$ è¡¨ç¤ºâ€œä¸ä¸‹é›¨â€

- $y = 1$ è¡¨ç¤ºâ€œå¸¦ä¼â€

äººä¸ºè®¾å®šï¼š

$$
P(y=1 \mid x=0)=0.2,\qquad P(y=1 \mid x=1)=0.9
$$


## 5.1 éšæœºç”Ÿæˆæ•°æ®

```python
def sample_xy(N=2000):
    xs = np.random.randint(0,2,size=N)
    ys = []
    for x in xs:
        if x == 0:
            ys.append(np.random.rand() < 0.2)
        else:
            ys.append(np.random.rand() < 0.9)
    return np.array(xs), np.array(ys)

xs, ys = sample_xy()

# ç»éªŒä¼°è®¡ p(y|x)
p_y1_x0 = ys[xs==0].mean()
p_y1_x1 = ys[xs==1].mean()
p_y1_x0, p_y1_x1

```

# ğŸ§ª è´å¶æ–¯å®éªŒï¼šåŸºäºåŒ»ç–—æ£€æµ‹æ•°æ®çš„å…ˆéªŒâ€“åéªŒåˆ†æ

æœ¬å®éªŒé€šè¿‡ä¸€ä¸ªç®€åŒ–çš„**åŒ»ç–—æ£€æµ‹åœºæ™¯**ï¼Œç”¨ Python å®Œæˆä¸€æ¬¡å®Œæ•´çš„è´å¶æ–¯åˆ†æï¼Œé‡ç‚¹ç†è§£ï¼š

- å…ˆéªŒæ¦‚ç‡ï¼ˆpriorï¼‰

- åéªŒæ¦‚ç‡ï¼ˆposteriorï¼‰
- åŸºå‡†ç‡ï¼ˆbase rateï¼‰
- ä¿¡æ¯æ›´æ–° $P(D \mid T)$

> è®°å·è¯´æ˜ï¼š

```python
# 0. å¯¼å…¥ä¾èµ–
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ä¸ºäº†å®éªŒå¯é‡å¤
np.random.seed(42)
```

## 0ï¸âƒ£ ç”Ÿæˆä¸€ä»½â€œæ¨¡æ‹ŸçœŸå®â€çš„åŒ»ç–—æ£€æµ‹æ•°æ®

æˆ‘ä»¬å‡è®¾**è‡ªå·±æ˜¯ä¸Šå¸ï¼Œä¸»å®°ä¸€åˆ‡**å› æ­¤å¯ä»¥è®¾å®šï¼š

- æ€»æ ·æœ¬é‡ $N = 400$
- çœŸæ­£çš„æ‚£ç—…ç‡ï¼ˆæ€»ä½“å±‚é¢ï¼‰ä¸º $P(D=1) = 0.1$
- æ£€æµ‹æ–¹æ³•çš„çµæ•åº¦ï¼ˆæ•æ„Ÿåº¦ï¼‰ä¸º $P(T=1 \mid D=1) = 0.9$
- æ£€æµ‹æ–¹æ³•çš„ç‰¹å¼‚åº¦ä¸º $P(T=0 \mid D=0) = 0.95$

ç”Ÿæˆæ•°æ®æ­¥éª¤ï¼š

1. æŒ‰ç…§æ‚£ç—…ç‡æ¨¡æ‹Ÿæ¯ä¸ªäººæ˜¯å¦æ‚£ç—…ã€‚  
2. å¯¹æ¯ä¸ªä¸ªä½“ï¼Œæ ¹æ®å…¶çœŸå®ç—…æƒ…å’Œæ£€æµ‹æ€§èƒ½ï¼ˆçµæ•åº¦ã€ç‰¹å¼‚åº¦ï¼‰ï¼Œæ¨¡æ‹Ÿæ£€æµ‹ç»“æœã€‚  
3. å¾—åˆ°ä¸€ä¸ªåŒ…å«ä¸¤åˆ—çš„æ•°æ®é›†ï¼š
   - `disease`ï¼š0/1  
   - `test`ï¼š0/1  


```python
# 1. ç”Ÿæˆæ¨¡æ‹Ÿæ•°æ®

N = 400  # æ ·æœ¬é‡

# çœŸå®å‚æ•°ï¼ˆç”¨äºç”Ÿæˆæ•°æ®ï¼‰
p_disease_true = 0.10      # çœŸæ­£æ‚£ç—…ç‡ P(D=1)
sens_true = 0.90           # çµæ•åº¦ P(T=1|D=1)
spec_true = 0.95           # ç‰¹å¼‚åº¦ P(T=0|D=0)

# 1) æ¨¡æ‹Ÿæ˜¯å¦æ‚£ç—…
disease = (np.random.rand(N) < p_disease_true).astype(int)  # 0/1

# 2) æ¨¡æ‹Ÿæ£€æµ‹ç»“æœ
test = np.zeros(N, dtype=int)

for i in range(N):
    if disease[i] == 1:
        # æ‚£ç—…è€…ï¼šä»¥çµæ•åº¦ä¸ºé˜³æ€§æ¦‚ç‡
        test[i] = int(np.random.rand() < sens_true)
    else:
        # æœªæ‚£ç—…è€…ï¼šä»¥å‡é˜³æ€§ç‡(1-spec_true)ä¸ºé˜³æ€§æ¦‚ç‡
        test[i] = int(np.random.rand() < (1 - spec_true))

# 3) ç»„è£…æˆ DataFrame
df = pd.DataFrame({
    "disease": disease,
    "test": test
})

df.head(100)
```

```python
# 2. æ··æ·†çŸ©é˜µç»Ÿè®¡

# ä½¿ç”¨ pandas crosstab
ct = pd.crosstab(df["test"], df["disease"], rownames=["T"], colnames=["D"])
ct

# ä¸ºäº†åç»­æ–¹ä¾¿ï¼Œå–å‡º a,b,c,d
# æŒ‰ç…§è¡¨æ ¼å®šä¹‰ï¼š
# a = T=1 ä¸” D=1
# b = T=1 ä¸” D=0
# c = T=0 ä¸” D=1
# d = T=0 ä¸” D=0

# a = 
# b = 
# c = 
# d = 

N_check = a + b + c + d

a, b, c, d, N_check

```
