## 0. å®éªŒå‡†å¤‡

æœ¬å®éªŒä½¿ç”¨çº¿æ€§å›å½’çš„å‡æ–¹è¯¯å·®ï¼ˆMSEï¼‰ä½œä¸ºç›®æ ‡å‡½æ•°ï¼š

$$
L(w) = \frac{1}{N}\sum_{i=1}^{N} (x_i^T w - y_i)^2
$$

å…¶ä¸­ $w \in \mathbb{R}^d$ ä¸ºæ¨¡å‹å‚æ•°ã€‚

æˆ‘ä»¬å°†å®ç°ï¼š

- Batch Gradient Descentï¼ˆBatch GDï¼‰
- Stochastic Gradient Descentï¼ˆSGDï¼Œbatch size=1ï¼‰
- Mini-batch SGDï¼ˆbatch size > 1ï¼‰

å¹¶åœ¨â€œ**ç›¸åŒè®¡ç®—é¢„ç®—ï¼ˆç´¯è®¡æ ·æœ¬æ¢¯åº¦è®¡ç®—æ¬¡æ•°ï¼‰**â€ä¸‹æ¯”è¾ƒå®ƒä»¬çš„è¡¨ç°ã€‚


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams

# ä¸­æ–‡å­—ä½“ï¼ˆå¦‚ä½ çš„ç¯å¢ƒæ²¡æœ‰è¿™äº›å­—ä½“ï¼Œå¯åˆ æ‰è¿™ä¸¤è¡Œï¼‰
rcParams['font.sans-serif'] = ['Microsoft YaHei', 'SimHei']
rcParams['axes.unicode_minus'] = False

np.random.seed(0)
```

## 1. æ„é€ æ•°æ®ï¼ˆçº¿æ€§å›å½’ï¼‰

æˆ‘ä»¬æ„é€ ä¸€ä¸ªä¸­ç­‰è§„æ¨¡çš„æ•°æ®é›†ï¼š

- æ ·æœ¬æ•°ï¼š$N=5000$
- ç‰¹å¾ç»´åº¦ï¼š$d=10$
- æ ‡ç­¾ï¼š$y = X w_{\text{true}} + \epsilon$ï¼Œå…¶ä¸­ $\epsilon$ ä¸ºé«˜æ–¯å™ªå£°

ä½ å¯ä»¥è‡ªç”±è°ƒæ•´ $N$ã€å™ªå£°å¼ºåº¦ç­‰å‚æ•°ï¼Œä½†å»ºè®®ä¸è¦å¤ªå¤§ï¼Œä»¥å…è¿è¡Œè¿‡æ…¢ã€‚


```python
N = 5000
d = 10

X = np.random.randn(N, d)
w_true = np.random.randn(d)
noise = 0.5 * np.random.randn(N)
y = X @ w_true + noise

def loss(w):
    """MSE æŸå¤±"""
    r = X @ w - y
    return np.mean(r**2)
```

## 2. å®ç°æ¢¯åº¦ï¼šå…¨é‡ / å•æ ·æœ¬ / å°æ‰¹é‡

- Batch GD ä½¿ç”¨å…¨é‡æ¢¯åº¦ï¼š
  $$
  \nabla L(w) = \frac{2}{N} X^T (Xw - y)
  $$

- SGD / Mini-batch ä½¿ç”¨å­é›†è¿‘ä¼¼æ¢¯åº¦ï¼š
  $$
  \nabla L_{\mathcal{B}}(w) = \frac{2}{|\mathcal{B}|} X_\mathcal{B}^T (X_\mathcal{B}w - y_\mathcal{B})
  $$


```python
def grad_full(w):
    """å…¨é‡æ¢¯åº¦"""
    r = X @ w - y
    return 2 * X.T @ r / N

def grad_minibatch(w, idx):
    """å°æ‰¹é‡æ¢¯åº¦ï¼ˆidx æ˜¯æ ·æœ¬ä¸‹æ ‡æ•°ç»„ï¼‰"""
    Xi = X[idx]
    yi = y[idx]
    r = Xi @ w - yi
    return 2 * Xi.T @ r / len(idx)

def grad_single(w, i):
    """å•æ ·æœ¬æ¢¯åº¦ï¼ˆSGDï¼‰"""
    xi = X[i]
    yi = y[i]
    return 2 * xi * (xi @ w - yi)
```

### ä½ éœ€è¦å›ç­”

- å“ªæ¡æ›²çº¿æ›´å¹³æ»‘ï¼Ÿä¸ºä»€ä¹ˆï¼Ÿ

```python
def batch_gd(w0, lr, steps):
    w = w0.copy()
    losses = []
    for _ in range(steps):
        w -= lr * grad_full(w)
        losses.append(loss(w))
    return w, np.array(losses)

def sgd(w0, lr, steps):
#ä»¿ç…§batch_gdå‡½æ•°å®Œæˆ


w0 = np.zeros(d)
lr = 0.04

bg_w, bg_losses = batch_gd(w0, lr, steps=300)
sgd_w, sgd_losses = sgd(w0, lr, steps=300)

plt.figure(figsize=(8,5))
plt.plot(bg_losses, label="Batch GD")
plt.plot(sgd_losses, label="SGD (batch=1)")
plt.xlabel("è¿­ä»£æ­¥æ•°")
plt.ylabel("è®­ç»ƒæŸå¤±")
plt.title("é—®é¢˜1ï¼šBatch GD vs SGDï¼ˆæŒ‰è¿­ä»£æ­¥æ•°ï¼‰")
plt.legend()
plt.grid(True)
plt.show()
```

# âœ… é—®é¢˜ 3ï¼šåœ¨â€œç›¸åŒè®¡ç®—é¢„ç®—â€ä¸‹æ¯”è¾ƒ Batch GD ä¸ Mini-batch SGD

> è¿™ä¸€æ­¥æ˜¯ç†è§£ SGD ä¼˜åŠ¿çš„å…³é”®ï¼š  
> **æ¯”è¾ƒä¸èƒ½åªçœ‹è¿­ä»£æ¬¡æ•°ï¼Œè€Œè¦çœ‹è®¡ç®—ä»£ä»·ï¼ˆæ¢¯åº¦è®¡ç®—æ¬¡æ•°ï¼‰ã€‚**

### è®¡ç®—é¢„ç®—å®šä¹‰

- Batch GD æ¯åš 1 æ¬¡æ›´æ–°ï¼Œéœ€è¦è®¡ç®— $N$ ä¸ªæ ·æœ¬æ¢¯åº¦
- Mini-batch SGD æ¯åš 1 æ¬¡æ›´æ–°ï¼Œéœ€è¦è®¡ç®— $B$ ä¸ªæ ·æœ¬æ¢¯åº¦

æˆ‘ä»¬ç”¨â€œç´¯è®¡æ ·æœ¬æ¢¯åº¦è®¡ç®—æ¬¡æ•°â€ä½œä¸ºæ¨ªè½´è¿›è¡Œæ¯”è¾ƒã€‚

### è¦æ±‚

1. Batch GD åšå°‘é‡æ­¥ï¼ˆä¾‹å¦‚ 15 æ­¥ï¼‰
2. è®¡ç®—å…¶æ€»é¢„ç®—ï¼š$15 \times N$
3. åœ¨ç›¸åŒé¢„ç®—ä¸‹è¿è¡Œ Mini-batch SGDï¼ˆä¾‹å¦‚ batch size=32ï¼‰
4. ç”»å‡ºä¸¤è€…åœ¨â€œç›¸åŒé¢„ç®—æ¨ªè½´â€ä¸‹çš„æŸå¤±æ›²çº¿

### ä½ éœ€è¦å›ç­”

- åœ¨ç›¸åŒè®¡ç®—é¢„ç®—ä¸‹ï¼Œè°ä¸‹é™æ›´å¿«ï¼Ÿä¸ºä»€ä¹ˆï¼Ÿ


```python
def batch_gd_with_budget(w0, lr, steps):
    w = w0.copy()
    losses = []
    budget = []
    total = 0
    for _ in range(steps):
        w -= lr * grad_full(w)
        total += N
        budget.append(total)
        losses.append(loss(w))
    return np.array(budget), np.array(losses)

def minibatch_sgd_with_budget(w0, lr, batch_size, total_budget):
#ä»¿ç…§batch_gd_with_budget å®Œæˆ

w0 = np.zeros(d)
lr = 0.02

bg_budget, bg_losses = batch_gd_with_budget(w0, lr, steps=15)
total_budget = bg_budget[-1]

batch_size = 32
sgd_budget, sgd_losses = minibatch_sgd_with_budget(w0, lr, batch_size, total_budget)

plt.figure(figsize=(8,5))
plt.plot(bg_budget, bg_losses, 'o-', label="Batch GD")
plt.plot(sgd_budget, sgd_losses, '-', label=f"Mini-batch SGD (B={batch_size})")
plt.xlabel("ç´¯è®¡æ ·æœ¬æ¢¯åº¦è®¡ç®—æ¬¡æ•°ï¼ˆè®¡ç®—é¢„ç®—ï¼‰")
plt.ylabel("è®­ç»ƒæŸå¤±")
plt.title("é—®é¢˜3ï¼šç›¸åŒè®¡ç®—é¢„ç®—ä¸‹ Mini-batch SGD vs Batch GD")
plt.legend()
plt.grid(True)
plt.show()

print("Batch GD æ€»é¢„ç®— =", total_budget)
print("Mini-batch SGD å®é™…é¢„ç®— =", sgd_budget[-1])
```

- ä¸åŒéšæœºåˆå§‹ç‚¹æ˜¯å¦ä¼šæ”¶æ•›åˆ°ä¸åŒå±€éƒ¨æå°å€¼ï¼Ÿ
- è¿™è¯´æ˜â€œéšæœºåˆå§‹åŒ–â€åœ¨ä¼˜åŒ–ä¸­èµ·ä»€ä¹ˆä½œç”¨ï¼Ÿ


```python
def gd_1d(x0, eta=0.05, steps=200):
    x = float(x0)
    traj = [x]
    #è¡¥å……ä»£ç 
    return np.array(traj)

# å¤šæ¬¡éšæœºåˆå§‹åŒ–
M = 5
x0_list = np.random.uniform(-4, 4, size=M)

final_x = []
final_fx = []

for x0 in x0_list:
    traj = gd_1d(x0, eta=0.05, steps=250)
    x_end = traj[-1]
    final_x.append(x_end)
    final_fx.append(f(x_end))

final_x = np.array(final_x)
final_fx = np.array(final_fx)

# å¯è§†åŒ–ï¼šå‡½æ•°æ›²çº¿ + æ”¶æ•›ç‚¹
plt.figure(figsize=(8,4))
plt.plot(xs, f(xs), label="f(x)")
plt.scatter(final_x, final_fx, marker='o', label="ä¸åŒéšæœºåˆå§‹åŒ–çš„æ”¶æ•›ç‚¹")
plt.title("é—®é¢˜1ï¼šéšæœºåˆå§‹åŒ–ä¼šå¯¼è‡´ GD æ”¶æ•›åˆ°ä¸åŒå±€éƒ¨æå°å€¼")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()

print("æ”¶æ•›ç‚¹ï¼ˆå»é‡åç²—ç•¥ç»Ÿè®¡ï¼‰ï¼š")
# ç®€å•èšç±»ï¼šæŒ‰å°é˜ˆå€¼åˆå¹¶ï¼ˆè¯¾å ‚æ¼”ç¤ºè¶³å¤Ÿï¼‰
clusters = []
for x in np.sort(final_x):
    if not clusters or abs(x - clusters[-1]) > 0.2:
        clusters.append(x)
print("å¤§è‡´æ”¶æ•›ä½ç½®æ•°é‡ =", len(clusters))
print("å¤§è‡´æ”¶æ•›ä½ç½® =", np.round(clusters, 3))
```

# âœ… é—®é¢˜ 2ï¼šåŠ å…¥å™ªå£°çš„æ¢¯åº¦ä¸‹é™ï¼ˆNoisy GDï¼‰èƒ½å¦æ›´å®¹æ˜“â€œè·³å‡ºå±€éƒ¨æå°å€¼â€ï¼Ÿ

### èƒŒæ™¯

æ™®é€š GD ä¸€æ—¦è¿›å…¥æŸä¸ªå±€éƒ¨æå°å€¼é™„è¿‘ï¼Œé€šå¸¸ä¼šç¨³å®šæ”¶æ•›ï¼Œå¾ˆéš¾ç¦»å¼€ã€‚  
æˆ‘ä»¬è€ƒè™‘åœ¨æ›´æ–°ä¸­åŠ å…¥éšæœºå™ªå£°ï¼š

$$
x_{k+1} = x_k - \eta f'(x_k) + \epsilon_k,
\quad \epsilon_k \sim \mathcal{N}(0, \sigma^2)
$$

### ä»»åŠ¡

1. ä»åŒä¸€ä¸ªåˆå§‹ç‚¹ $x_0$ å‡ºå‘ï¼ˆå»ºè®®å–ä¸€ä¸ªä¼šè½å…¥â€œè¾ƒå·®å±€éƒ¨æå°å€¼â€çš„ç‚¹ï¼‰
2. åˆ†åˆ«è¿è¡Œï¼š
   - æ™®é€š GDï¼ˆæ— å™ªå£°ï¼‰
   - Noisy GDï¼ˆæœ‰å™ªå£°ï¼‰
3. å¯¹ Noisy GD é‡å¤å¤šæ¬¡ï¼ˆå› ä¸ºå®ƒæ˜¯éšæœºçš„ï¼‰
4. æ¯”è¾ƒæœ€ç»ˆæ‰¾åˆ°çš„æœ€å°å€¼æ˜¯å¦æ›´å¥½ï¼ˆå³æœ€ç»ˆ $f(x)$ æ˜¯å¦æ›´ä½ï¼‰

### ä½ éœ€è¦å›ç­”

- åŠ å™ªå£°åç»“æœæ˜¯å¦æ›´â€œåˆ†æ•£â€ï¼Ÿä¸ºä»€ä¹ˆï¼Ÿ
- sigmaå¢å¤§æˆ–å‡å°ï¼Œå¯¹ä¼˜åŒ–è¿‡ç¨‹æœ‰ä»€ä¹ˆå½±å“ï¼Ÿ
- æ˜¯å¦æ›´å®¹æ˜“å¾—åˆ°æ›´ä½çš„ $f(x)$ï¼Ÿå™ªå£°å¤ªå¤§/å¤ªå°ä¼šæ€æ ·ï¼Ÿ


```python
def noisy_gd_1d(x0, eta=0.05, sigma=0.05, steps=250):
    x = float(x0)
    traj = [x]
    #è¡¥å……ä»£ç 
    return np.array(traj)

# é€‰æ‹©ä¸€ä¸ªå›ºå®šåˆå§‹ç‚¹ï¼ˆä½ ä¹Ÿå¯ä»¥æ”¹ï¼‰
x0 = 2.8

traj_gd = gd_1d(x0, eta=0.05, steps=250)

# å¤šæ¬¡è¿è¡Œ noisy GD
R = 5
sigma = 0.2
noisy_finals = []
noisy_trajs = []

for _ in range(R):
    traj = noisy_gd_1d(x0, eta=0.05, sigma=sigma, steps=250)
    noisy_trajs.append(traj)
    noisy_finals.append((traj[-1], f(traj[-1])))

noisy_finals = np.array(noisy_finals, dtype=float)

# ç”»ä¸€æ¡æ™®é€š GD è½¨è¿¹ + å¤šæ¡ noisy GD è½¨è¿¹ï¼ˆåªç”»è½¨è¿¹ç‚¹åœ¨å‡½æ•°ä¸Šçš„æŠ•å½±ï¼‰
plt.figure(figsize=(8,4))
plt.plot(xs, f(xs), label="f(x)")

plt.plot(traj_gd, f(traj_gd), '-', linewidth=2, label="æ™®é€š GD è½¨è¿¹")

for traj in noisy_trajs:
    plt.plot(traj, f(traj), alpha=0.35)

plt.title(f"é—®é¢˜2ï¼šNoisy GDï¼ˆsigma={sigma}ï¼‰å¤šæ¬¡è¿è¡Œçš„è½¨è¿¹å¯¹æ¯”")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()

print("æ™®é€š GD æœ€ç»ˆç‚¹ï¼šx =", traj_gd[-1], "f(x) =", f(traj_gd[-1]))
print("Noisy GD æœ€ç»ˆ f(x) çš„ç»Ÿè®¡ï¼š")
print("min =", noisy_finals[:,1].min())
print("mean=", noisy_finals[:,1].mean())
print("max =", noisy_finals[:,1].max())
```

# âœ… é—®é¢˜ 3ï¼šéšæœºæœç´¢ï¼ˆRandom Searchï¼‰ä½œä¸ºâ€œæ— æ¢¯åº¦â€åŸºçº¿æ–¹æ³•

### èƒŒæ™¯

éšæœºæœç´¢ä¸ä½¿ç”¨æ¢¯åº¦ï¼Œå®ƒçš„æ€æƒ³æ˜¯ï¼š

- åœ¨å…è®¸èŒƒå›´å†…ä¸æ–­éšæœºé‡‡æ ·å€™é€‰è§£ $x$
- è®°å½•å½“å‰è§è¿‡çš„æœ€å°å‡½æ•°å€¼

å®ƒå¾ˆâ€œç²—ç³™â€ï¼Œä½†ä½œä¸ºåŸºçº¿å¾ˆå¸¸ç”¨ã€‚

### ä»»åŠ¡

1. åœ¨åŒºé—´ $[-4, 4]$ å†…éšæœºé‡‡æ · $T$ æ¬¡
2. è®°å½•â€œåˆ°ç›®å‰ä¸ºæ­¢çš„æœ€å°å€¼â€éšé‡‡æ ·æ¬¡æ•°å˜åŒ–çš„æ›²çº¿
3. ä¸ä¸€æ¬¡æ™®é€š GD çš„æœ€ç»ˆç»“æœè¿›è¡Œå¯¹æ¯”ï¼ˆè°æ›´å®¹æ˜“å¾—åˆ°æ›´ä½çš„ $f(x)$ï¼Ÿï¼‰

### ä½ éœ€è¦å›ç­”

- T çš„å€¼æ˜¯30å˜æˆ300ï¼Œæ˜¯å¦å¯ä»¥æå‡ä¼˜åŒ–æ•ˆæœï¼Ÿä¸ºä»€ä¹ˆï¼Ÿ
- éšæœºæœç´¢æ˜¯å¦æœ‰å¯èƒ½æ‰¾åˆ°å¾ˆå¥½çš„æå°å€¼ï¼Ÿå®ƒçš„ç¼ºç‚¹æ˜¯ä»€ä¹ˆï¼Ÿ
- åœ¨è¿™ä¸ªä¸€ç»´é—®é¢˜ä¸­ï¼Œä½ è§‰å¾—éšæœºæœç´¢å’Œæ¢¯åº¦æ³•è°æ›´åˆ’ç®—ï¼Ÿ


```python
def random_search(bounds=(-4,4), T=30):
    lo, hi = bounds
    best_x = None
    best_fx = np.inf
    best_curve = []
    for _ in range(T):
        x = np.random.uniform(lo, hi)
        fx = f(x)
        if fx < best_fx:
            best_fx = fx
            best_x = x
        best_curve.append(best_fx)
    return best_x, best_fx, np.array(best_curve)

# éšæœºæœç´¢
best_x, best_fx, curve = random_search(bounds=(-4,4), T=400)

# ç”¨åŒä¸€åˆå§‹ç‚¹è·‘ä¸€æ¬¡ GD åšå¯¹æ¯”
traj_gd2 = gd_1d(x0=-2, eta=0.05, steps=250)
gd_best_fx = f(traj_gd2[-1])

plt.figure(figsize=(8,4))
plt.plot(curve, label="éšæœºæœç´¢ï¼šå½“å‰æœ€ä¼˜ f(x)")
plt.axhline(gd_best_fx, linestyle='--', label="ä¸€æ¬¡ GD çš„æœ€ç»ˆ f(x)")
plt.xlabel("éšæœºé‡‡æ ·æ¬¡æ•°")
plt.ylabel("å½“å‰æœ€ä¼˜ f(x)")
plt.title("é—®é¢˜3ï¼šéšæœºæœç´¢ï¼ˆæ— æ¢¯åº¦ï¼‰ä½œä¸ºåŸºçº¿å¯¹æ¯”")
plt.legend()
plt.grid(True)
plt.show()

print("éšæœºæœç´¢æœ€ä¼˜ï¼šx =", best_x, "f(x) =", best_fx)
print("GDï¼ˆä» x0=0.5 å‡ºå‘ï¼‰æœ€ç»ˆï¼šx =", traj_gd2[-1], "f(x) =", gd_best_fx)
```

4. éšæœºæœç´¢çš„ä¼˜ç‚¹ä¸ç¼ºç‚¹å„æ˜¯ä»€ä¹ˆï¼Ÿå®ƒé€‚åˆä»€ä¹ˆåœºæ™¯ï¼Ÿ

ï¼ˆä¸éœ€è¦è¯æ˜ï¼Œåªè¦ç»“åˆä½ çœ‹åˆ°çš„å›¾åƒç°è±¡è§£é‡Šå³å¯ã€‚ï¼‰
